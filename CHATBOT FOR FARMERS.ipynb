# RANDOM FOREST ALGORITHIM
Random Forest is a popular machine learning algorithm that belongs to the supervised learning technique. It can be used for both Classification and Regression problems in ML. It is based on the concept of ensemble learning, which is a process of combining multiple classifiers to solve a complex problem and to improve the performance of the model.

As the name suggests, "Random Forest is a classifier that contains a number of decision trees on various subsets of the given dataset and takes the average to improve the predictive accuracy of that dataset." Instead of relying on one decision tree, the random forest takes the prediction from each tree and based on the majority votes of predictions, and it predicts the final output.

The greater number of trees in the forest leads to higher accuracy and prevents the problem of overfitting.

# Step involved in the working of random forest algorithim

**The following steps explain the working Random Forest Algorithm:**

**Step 1:** Select random samples from a given data or training set.

**Step 2:** This algorithm will construct a decision tree for every training data.

**Step 3:** Voting will take place by averaging the decision tree.

**Step 4:** Finally, select the most voted prediction result as the final prediction result.

```python
features = df[['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']]
target = df['label']
Xtrain, Xtest, Ytrain, Ytest = train_test_split(features, target, test_size=0.2, random_state=2)
RF.fit(Xtrain, Ytrain)
```

# Step 4:Function to perform Prediction

```python
# Function to perform prediction
def predict_crop(data):
    data = np.array(data)
    prediction = RF.predict(data)
    return prediction[0]
```

Here we defined a function predict_crop(data) that takes in a data parameter. Here's a breakdown of what the code does:

**1.data = np.array(data):** This line converts the data input into a NumPy array using np.array(). By converting it into an array, it ensures that the input is in a suitable format for making predictions.

**2.prediction = RF.predict(data):** This line uses the RandomForestClassifier model (RF) to make predictions on the input data. The predict() method of the model is called, passing in the data array as the input. It returns an array of predicted values.

**3.return prediction[0]:** This line returns the first element of the prediction array. Assuming here we are interested in a single prediction, it extracts the first prediction from the array and returns it as the output of the predict_crop() function.

at last, the predict_crop() function takes in a data array, converts it into a NumPy array, uses the trained RandomForestClassifier model (RF) to make predictions on the data, and returns the first prediction from the resulting array.

# Step 5: Function to handle user queries

```python
# Function to handle user queries
def chatbot_response(user_query):
    if user_query.lower() == "help":
        return "Please enter the values for N, P, K, temperature, humidity, pH, and rainfall to predict the crop."
    elif user_query.lower() == "exit":
        return "Thank you for using the Crop Recommendation System. Goodbye!"
    else:
        try:
            values = [float(x) for x in user_query.split()]
            if len(values) != 7:
                return "Please enter all the values for N, P, K, temperature, humidity, pH, and rainfall."
            prediction = predict_crop([values])
            return "Predicted Crop: " + str(prediction)
        except ValueError:
            return "Invalid input. Please enter numerical values for N, P, K, temperature, humidity, pH, and rainfall."
```

the code defines a **chatbot_response(user_query)** function that handles the user's queries and generates appropriate responses. Here's an overview of how the code works:

1.The function takes in a **user_query** parameter, which represents the input query from the user.

2.It checks if the user's query is **"help"** by comparing the lowercase version of user_query with "help". If it matches, the function returns a help message providing instructions on how to use the Crop Recommendation System.

3.It checks if the user's query is **"exit"** by comparing the lowercase version of user_query with "exit". If it matches, the function returns a farewell message to the user, indicating the end of their interaction with the Crop Recommendation System.

4.If the user's query is **neither "help" nor "exit"**, the function assumes it is a query with values for the crop prediction. It tries to parse the values from the user's query by splitting it into separate values and converting them to float using a list comprehension.

5.If the parsing is successful, it checks if the number of values is not equal to 7. If the condition is true, the function returns a message asking the user to provide all the necessary values for crop prediction.

6.If the number of values is equal to 7, it calls the **predict_crop()** function with the parsed values wrapped in a list. This function is assumed to be defined elsewhere and is responsible for making the actual crop prediction based on the provided values.

7.The function then returns a response indicating the predicted crop along with the predicted value obtained from the predict_crop() function.

8.If any error occurs during the process, such as when parsing the values or calling predict_crop(), an exception is caught (ValueError), and an appropriate error message is returned to the user.

In summary, this code implements a basic chatbot that handles user queries for the Crop Recommendation System. It provides instructions for usage, allows the user to exit the system, accepts user input for crop prediction values, calls a separate function to make the crop prediction, and returns the predicted crop as a response.

# Step 7: CHATBOT

```python
# Start the chatbot
print("Crop Recommendation System Chatbot")
print("Enter 'help' for assistance or 'exit' to quit.")

while True:
    user_input = input("User: ")
    response = chatbot_response(user_input)
    print("ChatBot:", response)
    if response == "Thank you for using the Crop Recommendation System. Goodbye!":
        break
```

The code sets up a basic interactive chatbot loop for the Crop Recommendation System. Here's how it works:

1.It prints the initial messages to greet the user and provide instructions on how to use the chatbot.

2.The code enters a while loop that continues indefinitely until the user decides to exit.

3.Inside the loop, it prompts the user for input using the input() function and stores the user's input in the user_input variable.

4.It then calls the **chatbot_response()** function with the **user_input** as an argument to generate a response.

5.The generated response is printed as the bot's message using the print() function.

6.It checks if the response is the farewell message, indicating that the user wants to exit. If it matches the farewell message, the loop is broken, and the program terminates.

By implementing this loop, the chatbot repeatedly interacts with the user, generating responses based on their input until the user chooses to exit by entering "exit" or a similar command.
